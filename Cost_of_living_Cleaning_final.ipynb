```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load dataset from an csv file

cost = pd.read_csv('cost-of-living.csv')
cost
```

```python
# Original columns from the dataset
cost.columns
```

```python
# dropping columns that we consider not relevant such as Cigarettes, Taxi, Mortgage...  
cost.drop(['Unnamed: 0','x27', 'x28', 'x30', 'x31', 'x32', 'x33', 'x34', 'x35', 'x37', 'x40', 'x43', 'x55', 'data_quality'], axis=1, inplace=True)
```

```python
### Restaurants 'x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8'

cost['restaurant'] = cost[['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8']].mean(axis=1)
cost.drop(['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8'], axis= 1, inplace= True)
```

```python
### Groceries x9, x10, x11, x12, x13, x14, x15, x16, x17, x18, x19, x20, x21, x22, x23, x24, x25, x26

cost['groceries'] = cost[['x9', 'x10', 'x11', 'x12', 'x13', 'x14', 'x15', 'x16', 'x17', 'x18', 'x19', 'x20',
                         'x21', 'x22', 'x23', 'x24', 'x25', 'x26']].sum(axis=1)
cost.drop(['x9', 'x10', 'x11', 'x12', 'x13', 'x14', 'x15', 'x16', 'x17', 'x18', 'x19', 'x20',
                         'x21', 'x22', 'x23', 'x24', 'x25', 'x26'], axis= 1, inplace= True)
```

```python
### Transport x29

cost['transport'] = cost[['x29']].mean(axis=1)
cost.drop(['x29'], axis= 1, inplace = True)
```

```python
### Freetime x39, x41

cost['freetime'] = cost[['x39', 'x41']].mean(axis=1)
cost.drop(['x39', 'x41'], axis= 1, inplace = True)
```

```python
### Utilities x36, x38

cost['utilities'] = cost[['x36', 'x38']].sum(axis=1)
cost.drop(['x36', 'x38'], axis= 1, inplace = True)
```

```python
### Childcare x42

cost['childcare'] = cost['x42']
cost.drop(['x42'], axis= 1, inplace = True)
```

```python
### Clothing

cost['clothing'] = cost[['x44', 'x45', 'x46','x47']].mean(axis=1)
cost.drop(['x44', 'x45', 'x46','x47'], axis= 1, inplace = True)
```

```python
### Rent x48, x49, x50, x51

cost['rent'] = cost[['x48', 'x49', 'x50', 'x51']].mean(axis=1)
cost.drop(['x48', 'x49', 'x50', 'x51'], axis= 1, inplace= True)
```

```python
### Rent x52, x53

cost['housing_price'] = cost[['x52', 'x53']].mean(axis=1)
cost.drop(['x52', 'x53'], axis= 1, inplace= True)
```

```python
### Renaming the column x54 

cost['average_salary'] = cost['x54']
cost.drop(['x54'], axis= 1, inplace = True)
```

```python
def filter_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_filter = Q1 - 1.5 * IQR
    upper_filter = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_filter) & (df[column] <= upper_filter)]

# List of numeric columns to filter
numeric_columns = ['restaurant', 'groceries', 'transport', 'freetime', 'utilities', 'childcare', 'clothing', 'rent', 'housing_price', 'average_salary']

# Apply the filter_outliers_iqr function to all numeric columns
cost_filtered = cost.copy()  # Create a copy of the original DataFrame

for column in numeric_columns:
    cost_filtered_2 = filter_outliers_iqr(cost_filtered, column)

    
cost_filtered_2.describe().round(2)
```

```python
def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_filter = Q1 - 1.5 * IQR
    upper_filter = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_filter) | (df[column] > upper_filter)]
    
    if not outliers.empty:
        print(f"Extreme values found in column {column}:")
        print(outliers[[column]])
        print("\n")
    
    return outliers

# List of numeric columns to check for outliers
numeric_columns = ['restaurant', 'groceries', 'transport', 'freetime', 'utilities', 'childcare', 'clothing', 'rent', 'housing_price', 'average_salary']

# Identify outliers in all numeric columns
outliers_found = pd.DataFrame()
for column in numeric_columns:
    outliers_found = pd.concat([outliers_found, identify_outliers(cost, column)], axis=1)

# Display the resulting DataFrame with identified outliers
print(outliers_found)
```
